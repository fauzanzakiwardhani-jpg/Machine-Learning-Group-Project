```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
data = pd.read_csv('3A.tsv', sep='\t')
data.head(5)
```

Kode yang Anda berikan digunakan untuk memuat dataset bernama 3A.tsv menggunakan pandas dan menampilkan 5 baris pertama dari dataset menggunakan head(5). Fungsi pd.read_csv('3A.tsv', sep='\t') membaca file TSV yang dipisahkan oleh tab dan mengonversinya menjadi sebuah DataFrame. Kemudian, data.head(5) digunakan untuk menampilkan 5 baris pertama dari dataset sebagai preview untuk melihat struktur dan sampel data yang ada dalam file tersebut.

```python
# Function to clean data and handle anomalies
def clean_and_handle_anomalies(df):
    """Handles missing values, invalid data types, and outliers."""
    # Fill missing categorical columns with 'Unknown'
    df[['Name', 'Genre', 'Publisher']] = df[['Name', 'Genre', 'Publisher']].fillna('Unknown')

    # Handle missing numerical columns: 'User_Score' and 'Critic_Score'
    df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
    df['User_Score'].fillna(df['User_Score'].mean(), inplace=True)
    df['Critic_Score'].fillna(df['Critic_Score'].mean(), inplace=True)

    # Set negative values in 'User_Count' to 0
    df['User_Count'] = df['User_Count'].apply(lambda x: max(x, 0))

    # Handle missing 'Critic_Count'
    df['Critic_Count'].fillna(0, inplace=True)

    return df

# Apply cleaning function
data = clean_and_handle_anomalies(data)
```

Fungsi clean_and_handle_anomalies digunakan untuk membersihkan data dengan menangani beberapa masalah umum seperti nilai yang hilang, tipe data yang tidak valid, dan nilai yang tidak wajar (outliers). Fungsi ini mengisi nilai yang hilang pada kolom kategorikal seperti Name, Genre, dan Publisher dengan 'Unknown'. Untuk kolom numerik seperti User_Score dan Critic_Score, fungsi ini mengonversi nilai yang tidak dapat diubah menjadi angka menjadi NaN dan mengganti nilai tersebut dengan rata-rata dari kolom yang bersangkutan. Selain itu, nilai negatif pada kolom User_Count akan diubah menjadi 0, dan nilai yang hilang pada kolom Critic_Count akan diisi dengan 0. Fungsi ini memastikan bahwa dataset siap digunakan dengan mengatasi masalah-masalah umum yang dapat muncul pada data mentah.

```python
# Exploratory Data Analysis (EDA)
def eda(df):
    """Performs EDA including missing values, summary statistics, and visualizations."""
    # Check for missing values and print count
    print("Missing values in each column:")
    print(df.isnull().sum())

    # Display summary statistics
    print("\nSummary Statistics:")
    print(df.describe())

    # Select only numeric columns for histograms and correlation matrix
    numeric_columns = df.select_dtypes(include=[np.number]).columns

    # Plot histograms for numeric columns
    df[numeric_columns].hist(bins=15, figsize=(12, 8), color='skyblue', edgecolor='black')
    plt.tight_layout()
    plt.show()

    # Correlation matrix heatmap
    if len(numeric_columns) > 1:
        correlation_matrix = df[numeric_columns].corr()
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
        plt.show()

# Apply EDA function
eda(data)
```

Fungsi eda di atas melakukan Exploratory Data Analysis (EDA) pada dataset dengan beberapa langkah penting untuk mengeksplorasi dan memahami data. Penjelasan langkah-langkahnya adalah sebagai berikut:
1. Mengecek Nilai yang Hilang: Fungsi ini pertama-tama memeriksa apakah ada nilai yang hilang pada setiap kolom dengan menggunakan isnull().sum(), dan kemudian menampilkan jumlah nilai yang hilang pada setiap kolom untuk memberi gambaran tentang kualitas data.
2. Menampilkan Statistik Deskriptif: Fungsi ini kemudian menampilkan statistik deskriptif dari dataset menggunakan describe(), yang mencakup informasi seperti mean, standar deviasi, nilai minimum, maksimum, dan kuartil dari kolom numerik.
3. Visualisasi Distribusi Data: Fungsi ini kemudian memilih kolom-kolom numerik dari dataset dan membuat histogram untuk masing-masing kolom tersebut. Histogram ini membantu untuk melihat distribusi data dari kolom numerik dengan jumlah bin yang ditentukan (15).
4. Visualisasi Matriks Korelasi: Jika ada lebih dari satu kolom numerik, fungsi ini juga menghitung dan menampilkan matriks korelasi antar kolom numerik menggunakan corr(), yang kemudian divisualisasikan dengan heatmap menggunakan seaborn untuk memudahkan interpretasi hubungan antar variabel.


```python
# Function for Content-based Recommender System (based on Genre)
def content_based_recommender(item_name, n_recommendations=5):
    """Recommends similar items based on the genre similarity using cosine similarity."""
    # Prepare Tfidf Vectorizer for genre similarity
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    genre_matrix = tfidf_vectorizer.fit_transform(data['Genre'].fillna(''))

    # Compute cosine similarity between all items
    cosine_sim_matrix = cosine_similarity(genre_matrix, genre_matrix)

    # Locate the index of the input item
    item_idx = data[data['Name'] == item_name].index
    if len(item_idx) == 0:
        return [f"'{item_name}' not found in the dataset."]

    item_idx = item_idx[0]

    # Get similarity scores and sort them in descending order
    similarity_scores = list(enumerate(cosine_sim_matrix[item_idx]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get the top N recommendations
    recommendations = [data['Name'].iloc[i[0]] for i in sorted_scores[1:n_recommendations+1]]
    return recommendations
```

Fungsi content_based_recommender adalah sistem rekomendasi berbasis konten yang memberikan saran item serupa berdasarkan kesamaan genre menggunakan cosine similarity. Fungsi ini pertama-tama membuat representasi vektor dari genre setiap item menggunakan TfidfVectorizer untuk mengukur seberapa mirip genre antar item. Kemudian, dihitung matriks cosine similarity untuk semua item dalam dataset. Setelah itu, fungsi mencari indeks item yang diminta, menghitung skor kesamaan dengan item lainnya, dan mengurutkannya untuk memberikan N rekomendasi teratas berdasarkan kesamaan genre. Jika item yang dicari tidak ditemukan, fungsi akan mengembalikan pesan kesalahan.








```python
# Evaluate the recommender function with different inputs
inputs = ["Wii Sports", "Super Mario Bros.", "Mario Kart Wii"]
for game in inputs:
    print(f"\nRecommendations for '{game}':")
    recommendations = content_based_recommender(game)
    for rec in recommendations:
        print(f"- {rec}")
```

Kode ini digunakan untuk mengevaluasi fungsi content_based_recommender dengan memberikan tiga input berbeda, yaitu "Wii Sports", "Super Mario Bros.", dan "Mario Kart Wii". Untuk setiap item dalam daftar inputs, kode ini akan memanggil fungsi rekomendasi dan mencetak hasil rekomendasi yang diberikan oleh sistem berdasarkan genre yang serupa. Fungsi content_based_recommender akan mencari item yang relevan dengan genre dari masing-masing game tersebut dan mengembalikan rekomendasi item lain yang memiliki genre yang mirip. Hasilnya akan ditampilkan dalam format yang menyebutkan rekomendasi game yang relevan untuk setiap input.

Hasil rekomendasi yang diberikan oleh sistem menunjukkan game-game yang memiliki kesamaan genre dengan input yang diberikan. Misalnya, untuk game Wii Sports, rekomendasi yang muncul meliputi game seperti Wii Sports Resort dan Wii Fit, yang juga bergenre olahraga dan kebugaran. Untuk Super Mario Bros., rekomendasi mencakup game lain dalam seri Mario seperti New Super Mario Bros. dan Super Mario World, yang juga berfokus pada petualangan platformer. Sementara itu, untuk Mario Kart Wii, game seperti Mario Kart DS dan Gran Turismo 3: A-Spec direkomendasikan, yang berhubungan dengan genre balapan. Rekomendasi ini dihasilkan berdasarkan kesamaan genre yang dihitung menggunakan cosine similarity antara genre setiap game input dengan genre game lainnya dalam dataset.
